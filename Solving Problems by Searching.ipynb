<h1>Problem formulation<span style="font-size:12px">[8][1]</span></h1>
Before we jump to the search algorithms, we first need to define and formulate the problem.
Problem formulation involves deciding what actions and states to consider, given the goal.  
because we need to abstract the state details.


**A problem can be defined formally by 5 components:**

 - ### Initial state:
 The Initial state that the agent starts. In our case, the initial state can be described as in: **S**
 
 - ### Actions:
  The possible actions available to the agent given a particular state. For example, ACTIONS (In: S) returns the set of action that can be executed in s. (In: S) {Go: A, Go: B}

 - ### Transition model:
 Thetransition modeldescribing what each action does. Let us represent it by RESULT(s,a) where s is the state the action is currently in and a is the action performed by the agent. In this example, RESULT (In: S, Go: A) = In: A.
 
 - ### State space:
  The set of all states which can be reached by sequence of actions, can be represented by a graph ( vertex : state and edge : action). In this example the state space is {(S,4),(A,2),(B,6),(C,2),(D,3),(G,0)}

 - ### Goal test:
 The goal test determining whether the current state is a goal state. In our case, the goal state is {In: G}

### Introduction of A* Algorithm
A* Search algorithm is one of the best techniques used in path-finding and It is fastest search algorithm. This algorithm uses combination of heuristic searching and searching based on the shortest path. And we use this algorithm to find the shortest path in real-life Situations.[5]

<br>
<h3 dir="rtl">مقدمة لخوارزمية <span dir="ltr">A*</span></h3>
<div dir="rtl">هذه الخوارزمية تعتبر من أفضل الخوارزميات للبحث عن أفضل مسار وتعد من أسرع خوارزميات البحث. تدمج هذه الخوارزمية طريقتين من طرق البحث طريقة البحث بالتكلفة المتوقعة وطريقة البحث بأقل تكلفة للمسار.تستخدم هذه الخوارزمية في الكثير من الأمثلة في حياتنا الواقعية.</div>

<a id='Explan_a'></a>
### Explanation
A* Search Algorithm At each step it picks the node that having the lowest value of ‘f’ which is equal to the sum of ‘g’ and ‘h’.<br> We define ‘g’ as the exact cost from the start node to the goal node and ‘h’ as the estimated movement cost to move from the start node to goal node.[5]

<h3 dir="rtl">توضيح</h3>
<div dir= "rtl">  خوارزمية A* تأخذ في كل خطوة العقدة التي تحتوي على أقل قيمة من 'f' وهي تساوي مجموع 'g' و 'h'.
    <div>
    'g' عبارة عن التكلفة الحقيقة للمسار بين عقدة البداية وعقدة الهدف و 'h' عبارة عن التكلفة المتوقعة من عقدة البداية إلى عقدة الهدف.

# Heuristic function
The heuristic function is a way to inform the search about the direction to a goal. It provides an informed way to guess which neighbor of a node will lead to a goal.
There is nothing magical about a heuristic function. It must use only information that can be readily obtained about a node. Typically a trade-off exists between the amount of work it takes to derive a heuristic value for a node and how accurately the heuristic value of a node measures the actual path cost from the node to a goal.[6]

 -  A* algorithm (for GRAPH-SEARCH) is optimal, provided that **h(n)consistent** .

### Implementation
<h3 dir="rtl">تطبيق الخوارزمية </h3>

```python
import copy
#Create class Node
class Node:
    def __init__(self, name, h, f=0, p=None, g=0):
        self.name = name
        self.h = h
        self.f = f
        self.p = p
        self.g = g
        
#Create class Graph
class Graph:
    nodes = {}
    edges = []
    edge_indices = {}
    
    def add_node(self, node):
        #check if the new node object is instance of class Node and isn't in nodes. 
        if isinstance(node, Node) and node.name not in self.nodes:
            self.nodes[node.name] = node
            #initialise the edges with zero
            for row in self.edges:
                row.append(0)
            self.edges.append([0] * (len(self.edges) + 1))
            self.edge_indices[node.name] = len(self.edge_indices)

    def add_edge(self, u, v, cost):
        #check if node1 and node2 is in nodes then add the value of edge between them in edges.
        if u in self.nodes and v in self.nodes:
            self.edges[self.edge_indices[u]][self.edge_indices[v]] = cost
            
    def A_star(self, s, g):
        #put the start node in open_list
        open_list = {s.name: s}
        closed_list = {}
        while len(open_list) > 0:
            min_value = float('inf')
            for k, v in open_list.items():
                #choseing the min value
                if v.f < min_value:
                    min_value = v.f
                    min_node = v
               #pop the min value from the open_list 
            min_node = open_list.pop(min_node.name)
              #if min value equal goal
            if min_node.name == g.name:
                #put min value in closed_list because the goal found
                closed_list[min_node.name] = min_node
                #stop loop because the goal found 
                break
            index = 0
            
                #loop around one row in edge_indices to find successors of min node

            for n in self.edges[self.edge_indices[min_node.name]]:
                #if n != 0 then is a successor 
                if n != 0:
                    #copy the successor in varible ver
                    successor_copy = copy.deepcopy(self.nodes[list(self.edge_indices.keys())[index]])
                    #calculate f for successor
                    successor_copy.f = n + successor_copy.h + min_node.g 
                    #calculate g for successor
                    successor_copy.g = n + min_node.g
                    #if successor not on open_list and not on closed_list then put the successor in open_list
                    if successor_copy.name not in list(closed_list.keys()) and successor_copy.name not in list(open_list.keys()):
                            open_list[successor_copy.name] = successor_copy
                            #determine the current parent of successor 
                            successor_copy.p = min_node
                     #if successor on open_list and the f of successor is smaller than the old value f of successor      
                    if successor_copy.name in list(open_list.keys())and successor_copy.f < open_list[successor_copy.name].f:
                        #update the values of successor 
                        open_list[successor_copy.name] = successor_copy
                        #determine the current parent of successor 
                        successor_copy.p = min_node
                index += 1
            #put the min value in the closed_list 
            closed_list[min_node.name] = min_node
            

        #method to find the path by finding the parent of each node
        def find_path(node, temp_list):
            temp_list.append(node.name)
            while node.p != None and node!=s :
                temp_list.append(node.p.name)
                node = node.p
            return temp_list[::-1]

        print('close list: ',closed_list.keys())
        print('cost: ', closed_list['G'].g)
        print('path = ', find_path(closed_list['G'], []))
```

<a id='intro_sa'></a>
### Introduction of Simulated Annealing Algorithm
 The simulated annealing algorithm was originally inspired from the process of annealing in metal work. Annealing involves heating and cooling a material to alter its physical properties due to the changes in its internal structure. As the metal cools its new structure becomes fixed, consequently causing the metal to retain its newly obtained properties.[3]
<h3 dir="rtl">مقدمة خوارزمية محاكاة التلدين  </h3>

### Explanation
In simulated annealing we keep a temperature variable to simulate this heating process. We initially set it high and then allow it to slowly 'cool' as the algorithm runs. While this temperature variable is high the algorithm will be allowed, with more frequency, to accept solutions that are worse than our current solution. This gives the algorithm the ability to jump out of any local optimums it finds itself in early on in execution. As the temperature is reduced so is the chance of accepting worse solutions, therefore allowing the algorithm to gradually focus in on a area of the search space in which hopefully, a close to optimum solution can be found. [3]

### Implementation
<h3 dir="rtl">تطبيق الخوارزمية </h3>

```python
import random
import math
#Create class Node
class Node:
    def __init__(self, name, h=0, p=None , g=0):
        self.name = name
        self.h = h
        self.p = p
        self.g = g
        
#Create class Graph
class Graph:
    nodes = {}
    edges = []
    edge_indices = {}
    def add_node(self, node):
        #check if the new node object is instance of class Node and isn't in nodes.
        if isinstance(node, Node):
            self.nodes[node.name] = node
            #initialise the edges with zero
            for row in self.edges:
                row.append(0)
            self.edges.append([0] * (len(self.edges) + 1))
            self.edge_indices[node.name] = len(self.edge_indices)

    def add_edge(self, u, v, cost):
        #check if node1 and node2 is in nodes then add the value of edge between them in edges.
        if u in self.nodes and v in self.nodes:
            self.edges[self.edge_indices[u]][self.edge_indices[v]] = cost

    def SimulatedAnnealing(self, start , goal): 
      # Set initial value of temperature ,alpha and current
        temperature =10
        alpha = 0.96
        current = start
        # loop until temperature ==0 or current is the goal
        while temperature >0 and current.name != goal.name :
            #cooling temperature
            temperature *= alpha
            #find temperature equal 0 stop loop
            if temperature == 0:
                break   
            index =random.randrange(len(self.nodes)) 
            cost = self.edges[self.edge_indices[current.name]][index]
            
            while cost == 0:
                #find random index 
                index =random.randrange(len(self.nodes)) 
                cost = self.edges[self.edge_indices[current.name]][index]
            #set random node th next_node   
            next_node = self.nodes[list(self.edge_indices.keys())[index]]
            #the different between current heuristics and  next_node heuristics set to delta_e
            delta_e= current.h - next_node.h
             #if the delta_e >0 or the e^(delta_e / temperature)< random number in the range [0.0, 1.0)
            if delta_e > 0 or math.exp(delta_e / temperature) < random.random():
                #calculate g for next_node
                next_node.g =cost+ current.g
                #determine the current parent of successor
                next_node.p = current
                #set next_node to the current 
                current = next_node
                
        #method to find the path by finding the parent of each node 
        def find_path(node, temp_list):
            temp_list.append(node.name)
            
            while node.p != None and node!=start :
                temp_list.append(node.p.name)
                node = node.p
            return temp_list[::-1]
        
        print('path = ',find_path(current,[]))
        print('cost = ',goal.g)
```
